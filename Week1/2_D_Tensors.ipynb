```python
A 2d tensor can be viewed as a container the holds numerical values of the same type. 
In 2D tensors are essentially a matrix. Each row is a different sample and each column is a feature or attribute. 
```

```python
We can also represent gray-scale images as 2D tensors. 
The image intensity values can be represented as numbers between 0 and 255. 0 corresponds to color black and 255 white.
```

```python
Tensors can be extended to any number of dimensions. A 3D tensor is a combination of 3 1D tensors. 
```


<br>
import torch<br>
import numpy<br>
import pandas<br>
#Lets create a 2-D tensor<br>
#We first create a list with 3 nested lists.<br>
a = [[11,12,13], [21,22,23], [31,32,33]]<br>
b = torch.tensor([[11,12,13], [21,22,23], [31,32,33]])<br>
#We then cast the list to a torch tensor<br>
A = torch.tensor(a)<br>
print(A)<br>
#Lets check the no. of dimensions or rank<br>
print("Numer of dimensions of A = ", A.ndimension())<br>
#The first list [] represents the first dimensions and the second represents the second dimension<br>
#2D Tensors is as follows: [[]]<br>
#Lets check the number of rows and columns of A. It should be 3,3 --- 3 rows, 3 columns<br>
print("Shape of tensor A: ", A.shape)<br>
#OR<br>
print("Shape of tensor A: ", A.size())<br>
#The 3,3 tensor has 2 axes. Axis = 0 (vertical) and Axis = 1 (Horizontal)<br>
#Number of elements in a tensor -- using numel() method<br>
print("Number of elements in A: ", A.numel())<br>
#Indexing and Slicing 2D Tensors<br>
#Indexing<br>
print(A)<br>
A[0][1] #Element in 1st row and 2nd column<br>
A[1][2] #Element in 2nd row and 3rd column<br>
A[2][0] #Element in 3rd row and first column<br>
#Slicing<br>
A[1:3,2] #Slicing elements in rows 2 and 3 from the 3rd column<br>
A[2,0:3] #Slicing all the elements in the 3rd row<br>
#Adding 2D tensors only works for tensors of the same type <br>
#Lets add A and B. Elements of the same position will be added<br>
B = torch.tensor([[11,12,13], [21,22,23], [31,32,33]])<br>
C = A + B<br>
C<br>
#Multiplication by a scalar is the same as multiplying a matrix by a scalr<br>
#Multiplication of tensors is an elemenet-wise multiplication. Same position elements<br>
D = A*B<br>
print(D)<br>
#Matrix multiplication can be done in torch but same rules will apply <br>
#First matrix must have equal columns to the rows of the second matrix<br>
A = torch.tensor([[0,1,1],[1,0,1]])<br>
B = torch.tensor([[1,1],[1,1],[-1,1]])<br>
#Matrix multiplication is done by using the mm method<br>
C = torch.mm(A,B)<br>
print(C)
