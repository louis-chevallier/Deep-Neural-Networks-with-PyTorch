1-D Tensors in Pytorch<br>
------------------------------------------------------------------------------<br>
Tenors are arrays that are the building blocks of a neural network. <br>
- A 0-D tenors is just a number: 1, 2, 0.2 , 10<br>
1-D tensor is an array of numbers. It can be a row in a database, a vector, time series etc...<br>
A tensor contains elements of a single data type. The Tensor Type is the type of tensor.<br>
For real numbers: Tensor type is either a float or a double tensor or even half tensor.<br>
32-bit float -                          Float Tensor<br>
64-bit float -                          Double Tensor<br>
16-bit float -                          Half Tensor<br>
8-bit int (unsigned)(8-bit images)-     Byte Tensor     <br>
8-bit int (signed)-                     Char Tensor<br>
16-bit int (singed)-                    Short Tensor<br>
32-bit int (singed)-                    Int Tensor<br>
64-bit int (signed)-                    Long Tensor<br>

e can convert a numpy array to torch tensor this way

```python
import numpy as np
numpy_array = np.array([0.0, 1.0, 2.0, 3.0, 4.0])
torch_tensor = torch.from_numpy(numpy_array)
torch_tensor
```

onverting torch tensor back to numpy array

```python
back_to_numpy = torch_tensor.numpy()
back_to_numpy
```

ector addition in pytorch<br>
ddition must be done with vectors of the same type

```python
u = torch.tensor([1.0, 5.0])
v = torch.tensor([0.0, 3.0])
z = u + v
z = z.type(torch.FloatTensor)
print(z)
```

ector multiplicaiton with a Scalar

```python
y = torch.tensor([1,2])
z = 2*y
print(z)
```

roduct of two tensors (element wise multiplication)

```python
u = torch.tensor([1, 2])
v = torch.tensor([3, 2])
z = u*v
print(z)
```

ot product - element-wise multiplication and summation

```python
y = torch.dot(u,v)
print(y) #y is still a tensor
```
