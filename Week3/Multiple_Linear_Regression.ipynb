Multiple Linear Regression<br>


```python
from torch import nn
import torch
torch.manual_seed(1)
```

etting weight and bias manually

```python
w = torch.tensor([[2.0], [3.0]], requires_grad=True)
b = torch.tensor([[1.0]], requires_grad=True)
print(w)
print(b)
```

efine Prediction Function

```python
def forward(x):
    yhat = torch.mm(x, w) + b #torch.mm is matrix multiplicaiton. torch.matmul can also be used here. 
    return yhat
```

```python
x = torch.tensor([[1.0, 2.0]])
print("Dimensions of x: ", x.ndimension())
```

```python
yhat = forward(x)
print("The result: ", yhat)
```

ample tensor X. Lets create a 2D X with 3 samples (or rows)

```python
X = torch.tensor([[1.0, 1.0], [1.0, 2.0], [1.0, 3.0]])
```

ake the prediction of X 

```python
yhat = forward(X)
print("The result: ", yhat)
```

reate linear_regression Class using nn.Module Class

```python
class linear_regression(nn.Module):
    
    # Constructor
    def __init__(self, input_size, output_size):
        super().__init__()
        self.linear = nn.Linear(input_size, output_size)
    
    # Prediction function
    def forward(self, x):
        yhat = self.linear(x)
        return yhat
 
#Create multiple linear regression model with 2 inputs and 1 output
model = linear_regression(2, 1)
```

rint model parameters (auto initialized)

```python
print("The parameters: ", list(model.parameters()))
```

rint model parameters

```python
print("The parameters: ", model.state_dict())
```

ake a prediction of x

```python
yhat = model(x)
print("The result: ", yhat)
```

ake a prediction of X

```python
yhat = model(X)
print("The result: ", yhat)
```
