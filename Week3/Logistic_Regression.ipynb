Logistic Regression <br>


```python
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
```

```python
torch.manual_seed(1)
```

reate a tensor ranging from -100 to 100

```python
z = torch.arange(-100,100,0.1).view(-1,1)
print("The tensor: ", z)
```

reate a sigmoid object from torch.nn

```python
sig = nn.Sigmoid()
```

alculate yhat using sigmoid object

```python
yhat = sig(z)
print("The prediction yhat: ", yhat)
```

lot results

```python
plt.plot(z.numpy(), yhat.numpy())
plt.xlabel('z')
plt.ylabel('yhat')
```

ets try the sigmoid model in torch

```python
yhat1 = torch.sigmoid(z) #Element-wise Sigmoid from torch
```

lot results

```python
plt.plot(z.numpy(), yhat1.numpy())
plt.xlabel('z')
plt.ylabel('yhat1')
```

he plots are identical.

Custom Logistic Regression Class using nn.Module<br>


```python
class logistic_regression(nn.Module):
    
    def __init__(self, n_inputs):
        super().__init__()
        self.linear = nn.Linear(n_inputs, 1)
        
    def forward(self, x):
        yhat = torch.sigmoid(self.linear(x))
        
        return yhat
```

reate x and X tensor with 1 input

```python
x = torch.tensor([[1.0]])
X = torch.tensor([[-100], [0], [100.0]])
print('x = ', x)
print('X = ', X)
```

```python
model = logistic_regression(1) #Initializing model that will predict 1 dimension
```

ets view the randomly intialized parameters

```python
print("list(model.parameters()):\n", list(model.parameters()))
print("\nmodel.state_dict():\n", model.state_dict())
```

redict x and X

```python
yhat = model(x)
yhat1 = model(X)
```

ets create a new model that takes 2 inputs

```python
model = logistic_regression(2)
```

reate x and X tensor with 2 inputs

```python
x = torch.tensor([[1.0, 2.0]])
X = torch.tensor([[-100, -100], [0.0, 0.0], [-100, 100]])
print('x = ', x)
print('X = ', X)
```

redict x and X with 2 inputs

```python
yhat = model(x)
print(yhat)
```
